#1. The implementation of multivariable regression is done using the datasets

#2. the datasets boston consist of various features namely data and the target

#3. the *data act as input where as the target act as the output*

#4. the data is complex and we have to use multivariable regression to find the best prediction.

#5. we have to develop relationship between a dependent variable (i.e. an outcome of interest given by info.target) and more than 1 independent variable(info.data). This is how multivarible regression works.



```python
#so lets import the datasets form sklearn
from sklearn import datasets
```

```python
#so the boston datasets is loaded from the sklearn and it is assingned to the entity named data
data=datasets.load_boston()
```

```python
#lets just observe the data
data
```

```python
#so we know the data act as the input having multiple feature and the target act as output
#also the keypoint to notice here is that the feature_names actually represent the attribute of data(input) on which the target is given, simply means the output is dependent based on the attribute and the names of the attributes is provided by the feature_name present in the datasets
x=data.data 
y=data.target
x.shape, y.shape, data.feature_names
```

```python
#so we need to split our data into training and testing sets
#lets say we split the data based on 80: 20 ratio.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#now lets implement the inbuilt model of LinearRegression and lets apply it into the training set whereas the testing set is used to predict the test data 
from sklearn.linear_model import LinearRegression
algo=LinearRegression()
algo.fit(x_train,y_train)
algo.score(x_train,y_train), algo.score(x_test,y_test)
```

```python
#so the score suggest that the model works efficiently with training as well as testing sets
```

```python
#lets get the coef_ and the intercept_ on which the linear function between independent variable and dependent variable(output) is formed.
algo.coef_, algo.intercept_
```

```python
#so lets get the y_predicted from the same algorithm
y_predicted=algo.predict(x_test)
y_predicted-y_test
```
